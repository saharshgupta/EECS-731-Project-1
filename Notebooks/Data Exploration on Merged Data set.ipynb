## Exploring the merged Fifa 2018 and Fifa 2019 dataset
In this notebook we load and explore two different publicly available fifa 2018 and fifa 2019 dataset from kaggle.<br>
Fifa 2018 - https://www.kaggle.com/thec03u5/fifa-18-demo-player-dataset <br>
Fifa 2019 - https://www.kaggle.com/karangadiya/fifa19 <br>

This analysis is done after merging the two given data sets.

## Importing all required Libraries
Pandas is used to read/load the dataset. <br>
Matplotlib is used to create graphs and visualize the relation within the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading the data
pandas.read_csv(filepath) reads the data and creates a dataframe<br>
Printing out the head to see the top rows

```python
path1 = r"..\Data\Fifa18.csv"
path2 = r"..\Data\Fifa19.csv"
df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)
df1.head(3)
```

```python
df2.head(3)
```

## Cleaning the Data
We see there is an Unnamed column which is same as index, we remove it as its a redundant feature. <br>
Printing out the head to see the updated data set.

```python
df1.drop("Unnamed: 0",axis=1,inplace=True)
df2.drop("Unnamed: 0",axis=1,inplace=True)
df1.head(3)
```

```python
df2.head(3)
```

## Adding feature
Added Feature called year to both the dataset, so that the data points are distinguishable after being merged.

```python
# Adding a feature so that the data points are distinguishable when merged
df1["Year"] = [2018 for i in range(len(df1))] # Adding year 2018 to the 2018 Fifa dataset
df2["Year"] = [2019 for i in range(len(df2))] # Adding year 2019 to the 2019 Fifa dataset
```

##  Mearging and Analysing the dataset
Printing the shape of the dataset to check how many datapoints and features do we have. <br>
Fifa 2018 Data frame (df1) Shape:  (17981, 74) <br>
Fifa 2019 Data frame (df2) Shape:  (18207, 88) <br>
Merged Dataframe (df) Shape:  (36188, 105) <br>
This data set has 36,188 data points or rows, and 105 features or columns. <br>
17 features were not common and so total features increased to 105.<br><br>
Printing out the names of columns to see what are the feature names.

```python
df = pd.concat([df1,df2],sort = True) # Concating the two data frames together.
# Removing the letter K from the end of numbers, converting string to numbers so that we can run analysis
df["Wage"] = df["Wage"].replace('[\€*K+,]', '', regex=True).astype(float)
df.sort_values(["Wage","Overall"],inplace=True,ascending=False) # Sorting the merged dataset with Wage and Overall as key features 
print("Merged Dataframe Shape: ",df.shape)
print(df.columns)
```

## Sorting and selecting key features
From the list features we see mostly all the features are important, different features can help us predict different things. Such as if we look at "Wage" and "Overall" we can tell how much a fifa player is paid depending on the "Overall" or "Potential" in the year 2019.<br><br>
With this data set we can form a team of best players (Like Classic XI) based on there stats given in the data, we can also predict how much a player gets paid based on different stats from the data. <br><br>
From the list of features I think the most major features are "Name", "Overall", "Potential", "Value", and "Wage", Wage is in string, so I stripped the euro sign and converted it to float to run analytics.<br><br>
Sorting the dataset on wage and overall, and printing out the selected features.
